# Check result accuracy

```python
uuidmap = {j['uuid']: j for j in df.to_dict('records')}

correct = sum(1 if uuidmap[r['uuid']]['spoilerType'] == r['spoilerType'] else 0 for r in results.to_dict('records'))
correct
```

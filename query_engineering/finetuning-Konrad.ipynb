```python
print('BEISPIEL VON ALEX/JULIUS')
import pandas as pd
def number_of_words(inputString):
    result = len(inputString.split())
    return result
df = pd.read_json('../data/validation.jsonl', lines=True)
df['targetParagraphs'] = df['targetParagraphs'].apply(lambda p: "\n".join(p))
df['spoiler_text_contains_numbers'] = df['targetParagraphs'].apply(lambda p: number_of_words(p))
print(df["spoiler_text_contains_numbers"])
print(df["targetParagraphs"])
for ix, targetParagraphs in df["targetParagraphs"].items():
    print(str(ix)+": "+str(targetParagraphs))
```
